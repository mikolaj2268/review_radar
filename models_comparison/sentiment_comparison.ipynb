## trenowanie modeli liniowych uwzgledniajac mozliwosc pomylki o 1 gwiazdke

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# ----------------------------
# 1. Data Preparation
# ----------------------------

# Assuming `df` is your dataset
df = df.copy()  # Create a copy of the dataset to ensure we don't modify the original

# Drop rows with missing values in the feature columns or target column
df = df.dropna(subset=[
    'textblob_pattern_polarity',
    'vader_neg', 'vader_neu', 'vader_pos', 'vader_compound',
    'distilbert_negative_prob', 'distilbert_positive_prob',
    'roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob',
    'score'
])

# Define the target variable
y = df['score']

# Split the dataset into training and testing subsets (80/20 split)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# ----------------------------
# 2. Model Training
# ----------------------------

# Initialize dictionaries to store results
models = {}
metrics = {}

# Helper function to compute tolerance accuracy
def tolerance_accuracy(y_true, y_pred, tolerance=1):
    y_pred_rounded = np.round(y_pred)  # Round predictions to nearest integer
    correct = np.abs(y_true - y_pred_rounded) <= tolerance
    return correct.mean()

# Helper function to train and evaluate a model
def train_and_evaluate_model(X_train, X_test, y_train, y_test, model_name):
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    tol_acc = tolerance_accuracy(y_test, y_pred)  # Compute tolerance accuracy
    return model, mae, r2, tol_acc

# 1. Model using `textblob_pattern_polarity`
X_train_1 = train_df[['textblob_pattern_polarity']]
X_test_1 = test_df[['textblob_pattern_polarity']]
models['textblob'], metrics['textblob_mae'], metrics['textblob_r2'], metrics['textblob_tol_acc'] = train_and_evaluate_model(
    X_train_1, X_test_1, train_df['score'], test_df['score'], "TextBlob"
)

# 2. Model using VADER features
X_train_2 = train_df[['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']]
X_test_2 = test_df[['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']]
models['vader'], metrics['vader_mae'], metrics['vader_r2'], metrics['vader_tol_acc'] = train_and_evaluate_model(
    X_train_2, X_test_2, train_df['score'], test_df['score'], "VADER"
)

# 3. Model using DistilBERT features
X_train_3 = train_df[['distilbert_negative_prob', 'distilbert_positive_prob']]
X_test_3 = test_df[['distilbert_negative_prob', 'distilbert_positive_prob']]
models['distilbert'], metrics['distilbert_mae'], metrics['distilbert_r2'], metrics['distilbert_tol_acc'] = train_and_evaluate_model(
    X_train_3, X_test_3, train_df['score'], test_df['score'], "DistilBERT"
)

# 4. Model using RoBERTa features
X_train_4 = train_df[['roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob']]
X_test_4 = test_df[['roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob']]
models['roberta'], metrics['roberta_mae'], metrics['roberta_r2'], metrics['roberta_tol_acc'] = train_and_evaluate_model(
    X_train_4, X_test_4, train_df['score'], test_df['score'], "RoBERTa"
)

# ----------------------------
# 3. Results Comparison
# ----------------------------

# Print out the results
print("Model Performance Metrics:")
for model_name in ['textblob', 'vader', 'distilbert', 'roberta']:
    print(f"{model_name.upper()}:")
    print(f"  - Mean Absolute Error (MAE): {metrics[f'{model_name}_mae']:.2f}")
    print(f"  - R-squared (R¬≤): {metrics[f'{model_name}_r2']:.2f}")
    print(f"  - Tolerance Accuracy (¬±1): {metrics[f'{model_name}_tol_acc']:.2%}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Funkcja do obliczenia korelacji grupowej
def group_correlation(X, y):
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)  # R-squared jako miara dopasowania
    return r2

# Wyb√≥r zestaw√≥w zmiennych
features_groups = {
    'TextBlob': ['textblob_pattern_polarity'],
    'VADER': ['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound'],
    'DistilBERT': ['distilbert_negative_prob', 'distilbert_positive_prob'],
    'RoBERTa': ['roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob']
}

# Obliczanie R-squared dla ka≈ºdej grupy
group_correlations = {}
for group_name, features in features_groups.items():
    r2 = group_correlation(df[features], df['score'])
    group_correlations[group_name] = r2

# Wy≈õwietlenie wynik√≥w
print("Korelacje grupowe (`score` od zmiennych grupy) jako R-squared:")
for group, r2 in group_correlations.items():
    print(f"{group}: {r2:.2f}")
```

## random forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Helper function to train and evaluate random forest models
def evaluate_random_forest(X, y, model_name):
    model = RandomForestRegressor(random_state=42)
    model.fit(X, y)
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    return mse

# Evaluate Random Forest for each sentiment model
rf_results = {
    'textblob': evaluate_random_forest(df[['textblob_pattern_polarity']], df['score'], "TextBlob"),
    'vader': evaluate_random_forest(df[['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound']], df['score'], "VADER"),
    'distilbert': evaluate_random_forest(df[['distilbert_negative_prob', 'distilbert_positive_prob']], df['score'], "DistilBERT"),
    'roberta': evaluate_random_forest(df[['roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob']], df['score'], "RoBERTa"),
}

# Print results
print("Random Forest MSE for each sentiment model:")
for model_name, mse in rf_results.items():
    print(f"{model_name}: {mse:.2f}")
```
