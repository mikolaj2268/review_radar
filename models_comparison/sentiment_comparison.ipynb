## Correlation

```python
correlation_features = [
    'textblob_pattern_polarity',
    'vader_neg', 'vader_neu', 'vader_pos', 'vader_compound',
    'distilbert_negative_prob', 'distilbert_positive_prob',
    'roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob',
    'score'
]

correlation_matrix = df[correlation_features].corr()

correlation_with_score = correlation_matrix['score'].drop('score')

print("Correlations of the `score` variable with features:")
print(correlation_with_score)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def group_correlation(X, y):
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    return r2

features_groups = {
    'TextBlob': ['textblob_pattern_polarity'],
    'VADER': ['vader_neg', 'vader_neu', 'vader_pos', 'vader_compound'],
    'DistilBERT': ['distilbert_negative_prob', 'distilbert_positive_prob'],
    'RoBERTa': ['roberta_negative_prob', 'roberta_neutral_prob', 'roberta_positive_prob']
}

group_correlations = {}
for group_name, features in features_groups.items():
    r2 = group_correlation(df[features], df['score'])
    group_correlations[group_name] = r2

print("Group correlations (`score` from group variables) as R-squared:")
for group, r2 in group_correlations.items():
    print(f"{group}: {r2:.2f}")
```
